```python
#import dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

UNIVARIATE ANALYSIS


```python
sns.histplot(df['price'])
```

We observed that the distribution of the car feature is positiveky skewed, which is not a good statistical value. 
Quite a number of the cars are cheap, the most expensive car falls within the range of 40,000 - 45,000 and they are not up to 10. 
Greater number of cars from this data falls within 5,000 - 20,000.

```python
#univariate feature of Fuel type
sns.countplot(x='fueltype', data = df)
```

We observed that the fueltype in use by most cars run on gas and we dont have up to 25 cars that run on diesel.


```python
#univariate distribution of engine type
sns.countplot(x='enginetype', data=df)
```

It is observed that ohc enginetype is the most used type of engine.

```python
#car body feature
sns.countplot(x='carbody', data = df)
```

A large trunk of the cars which are almost 100 have Sedan Crbody, followed by Hatchback which have about 70 cars.

BIVARIATE ANALYSIS

```python
#Relationship between horsepower and price
sns.jointplot(x='horsepower', y='price', data=df)
```

Horsepower is an important feature in defining the price of the cars, as the plo is showing us that those cars having very high horse powers are very expensive and vice versa.

```python
sns.scatterplot(x='horsepower', y='price', data=df)
```

We can observe a positive correlation between the horse power and our dependent variable.
I can say that horsepower feature is fairly symetrical though i can detect the presence of outliers also.

```python
#relationship between the cars engine location and car priced
sns.boxplot(x='enginelocation', y='price', data=df)
```

We can observe that cars that their prices falls between 7,000 to about 30,000 have their engines at the front.
While cars that have their engines at the rear are generally more expensive.

```python
sns.barplot(x='enginelocation', y='enginesize', data=df)
```

On an average, cars with bigger engine sixe are at the rear part of the car.

```python
sns.stripplot(x='enginetype', y='highwaympg', data=df)
```

People tend to go for ohc engine type because it has more highway mpg.
The higher the mpg, the more economical the car is.

```python
sns.boxplot(x='carbody', y='price', data=df)
```

We can observe that on average, hardtop and conversible cars are more expensive. While hatchback is the cheapest by far, followed by sedan and wagon.
